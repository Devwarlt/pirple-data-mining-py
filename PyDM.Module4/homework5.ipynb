## Real-life usage of matrix methods
By default, datasets provide a lot of useful information whose could be used to predict data and do statistic measurements. However, for some cenarios, we might use matrix operations to provide some, consecutively, more extra informations that could be used for data correlation. Furthermore, they can be meaningful to deduce ratings across correlated entries from dataset.

On current example, I'm using matrix transpose to change orientation of columns to rows and easily visualize each player's stats. But, it can also be referenced to solve an equation using linear algebra and vector transform to proceed with generation of values like **`Overall`**, **`Potential`** etc, as you can see on image below:
