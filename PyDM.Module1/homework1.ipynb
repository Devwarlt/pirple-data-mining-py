```python
"""

# DateTime conversion with Pandas:
    For each entry from data frame, we do convert to datetime type
    and then assign these entry with day of week name.

Note: since 'Series.dt.weekday_name' got deprecated on Pandas version
1.1.0, I had to switch to 'Series.dt.day_name()' instead.

References:
- "Num day to Name day with Pandas", available at
    https://stackoverflow.com/a/51795079;
- "What's new in 1.0.0 (January 29, 2020)", available at
    https://pandas.pydata.org/pandas-docs/stable/whatsnew/v1.0.0.html?highlight=weekday_name

"""
df['day_of_the_week'] = pd.to_datetime(df.date_of_listing).dt.day_name()
```
